```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from jupyterthemes import jtplot
jtplot.style()
%matplotlib inline
```

```python
data = pd.read_excel('world-population.xlsm')

df = pd.DataFrame(data, columns=['Year', 'Population'])

df.head()
```

```python
plt.plot(df['Year'], df['Population'], color='red')
plt.title('Population by Year')
plt.xlabel('Year')
plt.ylabel('Population (Billions)')
plt.grid(True)
```
